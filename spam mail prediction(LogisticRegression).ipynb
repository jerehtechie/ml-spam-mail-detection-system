```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_data=pd.read_csv(r"C:\Users\jerem\Desktop\ml data\mail_data.csv")
mail_data
```

```python
mail_data.isnull().sum()
```

```python
#replacing a null value with a null string
```

```python
mail_data=mail_data.where((pd.notnull(mail_data)),'')
```

```python
mail_data.shape
```

```python
#label encoding
```

```python
mail_data.loc[mail_data['Category']=='ham','Category']=1
mail_data.loc[mail_data['Category']=='spam','Category']=0
```

```python
mail_data
```

```python
#sepaerate data from the labels
```

```python
x=mail_data['Message']
print(x)
y=mail_data['Category']
print(y)
```

```python
#spliting the data into training set and test set
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
#feature extraction
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
training_features=feature_extraction.fit_transform(x_train)
testing_features=feature_extraction.transform(x_test)
```

```python
#converting y_train and y_test into intengers
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#training the model
```

```python
model=LogisticRegression()
model.fit(training_features,y_train)
```

```python
#evaluating our model
```

```python
train_prediction=model.predict(training_features)
training_accuracy=accuracy_score(train_prediction,y_train)
print(training_accuracy)
test_prediction=model.predict(testing_features)
test_accuracy=accuracy_score(test_prediction,y_test)
print(test_accuracy)
```

```python
#creating an spam mail detection system
```

```python
mail=["End-of-Year Learning Opportunities"]
mail_features=feature_extraction.transform(mail)
prediction=model.predict(mail_features)

if (prediction[0]==0):
    print('This email was identified as a spam')
elif(prediction[0]==1):
    print('This is ham mail and is save to open')
```
